```python
import numpy as np  # linear algebra
import pandas as pd  # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns  # data visualization
import matplotlib.pyplot as plt  # data visualization
import plotly.express as px  # interactive data visualization
import warnings  # to handle warnings
warnings.simplefilter('ignore')  # ignore warnings
```

```python
# Load the dataset
df = pd.read_csv('amazon.csv')

# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Get a summary of the dataset
print("\nDataset Summary:")
print(df.info())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['rating'], bins=10, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='discounted_price', y='rating')
plt.title('Discounted Price vs. Product Rating')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='category', y='rating', estimator=np.mean)
plt.title('Average Ratings by Product Category')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Download the VADER lexicon if you haven't done so
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer

# Initialize the sentiment intensity analyzer
sia = SentimentIntensityAnalyzer()

# Apply sentiment analysis to the review content
df['sentiment'] = df['review_content'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Visualize sentiment distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['sentiment'], bins=10, kde=True)
plt.title('Sentiment Distribution of Reviews')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Check data types
print(df.dtypes)

# Convert 'rating' to numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Now try recommending top products again
top_products = df.nlargest(10, 'rating')
print("Top 10 Recommended Products:")
print(top_products[['product_name', 'rating', 'discounted_price']])
```

```python
# Top Users by Review Count
top_users = df['user_name'].value_counts().head(10).reset_index()
top_users.columns = ['User Name', 'Review Count']
print("Top 10 Users by Review Count:")
from IPython.display import display
display(top_users)

# User Sentiment Analysis
user_sentiment = df.groupby('user_name')['sentiment'].mean().sort_values(ascending=False).head(10).reset_index()
user_sentiment.columns = ['User Name', 'Average Sentiment']
print("Average Sentiment by User:")
display(user_sentiment)

# User Ratings Analysis
user_ratings = df.groupby('user_name')['rating'].mean().sort_values(ascending=False).head(10).reset_index()
user_ratings.columns = ['User Name', 'Average Rating']
print("Average Rating by User:")
display(user_ratings)
```

```python
# Get the top 10 products by average rating
top_rated_products = df.groupby('product_name')['rating'].mean().nlargest(10)

# Convert the Series to a DataFrame for better formatting
top_rated_df = top_rated_products.reset_index()
top_rated_df.columns = ['Product Name', 'Average Rating']

# Display the DataFrame as a table
print("Top 10 Products by Average Rating:")
from IPython.display import display
display(top_rated_df)
```

```python
# Get the top 10 products by average sentiment
top_sentiment_products = df.groupby('product_name')['sentiment'].mean().sort_values(ascending=False).head(10)

# Convert the Series to a DataFrame for better formatting
top_sentiment_df = top_sentiment_products.reset_index()
top_sentiment_df.columns = ['Product Name', 'Average Sentiment']

# Display the DataFrame as a table
print("Top 10 Products by Average Sentiment:")
from IPython.display import display
display(top_sentiment_df)
```

```python
# Get the top 10 products by rating count
top_products = df.sort_values(by='rating_count', ascending=False).head(10)[['product_name', 'rating', 'rating_count']]

# Display the DataFrame as a table
print("Top 10 Products by Rating Count:")
from IPython.display import display
display(top_products)
```

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar=True, 
            linewidths=.5, linecolor='black')

# Add title
plt.title('Correlation Heatmap of Amazon Sales Dataset')

# Show the plot
plt.show()
```
